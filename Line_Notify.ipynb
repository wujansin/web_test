### YouBike Open Data


```python
url = "https://tcgbusfs.blob.core.windows.net/dotapp/youbike/v2/youbike_immediate.json"
data = requests.get(url).json()
data[0]
```

```python
x= data[0]
print(x['sna'], "-->", x['lat'], x['lng'])
```

```python
location = "台北市"
gps = geocoder.osm(location).latlng
m = folium.Map(gps, zoom_start=16)

for x in data:
    #print(x['sna'], "-->", x['lat'], x['lng'])
    gps = [x['lat'], x['lng']]
    addr = x['sna']
    folium.Marker(gps, popup = addr).add_to(m)
m
```

```python
m.save("index.html")
```
